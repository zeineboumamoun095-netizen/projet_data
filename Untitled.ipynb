```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import time
from sklearn.cluster import KMeans

#Q2
start_csv = time.time()

df = pd.read_csv("online_retail.csv")

end_csv = time.time()

print("Temps lecture csv:", end_csv - start_csv)

df.to_parquet("online_retail.parquet")

start_parquet = time.time()

df_parquet = pd.read_parquet("online_retail.parquet")

end_parquet = time.time()

print("Temps lecture Parquet:", end_parquet - start_parquet)

#Q3
manque = df_parquet.isnull().sum()

pourcent = (manque * 100) / len(df_parquet)
print(pourcent)

df_parquet = df_parquet.dropna(subset=['CustomerID'])

#Q4
df_parquet = df_parquet[df_parquet['Quantity'] > 0]
df_parquet = df_parquet[df_parquet['UnitPrice'] > 0]

#Q5
df_parquet['MontantTotal'] = df_parquet['Quantity'] * df_parquet['UnitPrice']

#Q6
df_parquet['InvoiceDate'] = pd.to_datetime(df_parquet['InvoiceDate'])

df_RFM = df_parquet.groupby('CustomerID').agg({
    'InvoiceDate' : lambda dates: (df_parquet['InvoiceDate'].max() - dates.max()).days,
    'InvoiceNo' : 'nunique',
    'MontantTotal' : 'sum'
})

df_RFM.columns = [['Recence','Frequence','Montant']]

#Q7
plt.hist(df_RFM['Montant'], bins=50)
plt.title("la variation du montant")

#Q8
plt.hist(df_RFM['Montant'], bins=50)
plt.title("la variation du montant")

#Q9
scaler = StandardScaler()
df_RFM_scaled = scaler.fit_transform(df_RFM)

#Q10
inertia = []
for k in range(1, 11):
   kmeans = KMeans(n_clusters=k, random_state=42)
   kmeans.fit(df_RFM_scaled)
   inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia)
plt.xlabel("k")
plt.ylabel("Inertia")
plt.title("Elbow Method")

#Q12
kmeans = KMeans(n_clusters=4, random_state=42)
df_RFM['Cluster'] = kmeans.fit_predict(df_RFM_scaled)

#Q13
#cluster_moyent = df_RFM.groupby('Cluster').mean()

#Q14
def classify_customer(row):

    if row['Recence'] < 50 and row['Frequence'] > 50:
        return 'Champions'

    elif row['Recence'] > 150:
        return 'At Risk'

    else:
        return 'Regular'


#cluster_analysis['Type'] = cluster_analysis.apply(classify_customer, axis=1)

#print(cluster_analysis)

#Q15
plt.scatter(
    df_RFM['Frequence'],
    df_RFM['Montant'],
    c=df_RFM['Cluster']
)

plt.xlabel("Frequence")

plt.ylabel("Montant")

plt.title("Customer Segments")

plt.show()
```
